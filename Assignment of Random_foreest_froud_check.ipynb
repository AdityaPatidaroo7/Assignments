```python
#Importing Librraries
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```python
#Load the dataset
dataset= pd.read_csv('Fraud_check (1).csv')
dataset.head()
```

```python
dataset2=dataset.rename({'Marital.Status':'MS','Taxable.Income':'Tax','City.Population':'CP','Work.Experience':'WP'},axis=1)
```

```python
dataset.columns
```

```python
dataset2.loc[dataset2.Tax>30000,'Taxincome']='norisk'
dataset2.loc[dataset2.Tax<=30000,'Taxincome']='riski'
```

```python
dataset2.drop('Tax',1,inplace=True)
```

```python
#Checking null Values
dataset2.isnull().sum()
```

```python
label_encoder = preprocessing.LabelEncoder() 
dataset2['Undergrad']= label_encoder.fit_transform(dataset2['Undergrad']) 
dataset2['MS']= label_encoder.fit_transform(dataset2['MS'])
dataset2['CP']= label_encoder.fit_transform(dataset2['CP'])
dataset2['Taxincome']= label_encoder.fit_transform(dataset2['Taxincome'])
dataset2['Urban']= label_encoder.fit_transform(dataset2['Urban'])
```

```python
dataset2
```

```python
colnames = list(dataset2.columns)
predictors = colnames[0:5]
target = colnames[5]
```

```python
predictors
```

```python
target
```

```python
#Deviding the data into X&Y as Target Variable And Predictors
X = dataset2[predictors]
Y = dataset2[target]
```

```python
X
```

```python
#Model building
rf = RandomForestClassifier(n_jobs=2,oob_score=True,n_estimators=1000,criterion="entropy")
```

```python
#fit the model
rf.fit(X,Y)
```

```python
rf.estimators_
```

```python
rf.classes_
```

```python
rf.n_features_
```

```python
rf.n_outputs_
```

```python
dataset2['rf_pred'] = rf.predict(X)
```

```python
#Confussin_Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(dataset2['Taxincome'],dataset2['rf_pred'])
```
